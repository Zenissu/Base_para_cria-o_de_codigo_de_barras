## 1 Introdução




O tema de escolha do grupo foi Geração/Leitura de Código de Barras ou QRCode. Em nosso projeto, realizamos a conversão de imagens para código de barras. O tema foi escolhido por conta de suas aplicações na indústria e no meio social e como esse código pode facilitar a utilização e criação de banco de dados.

O código de barras se trata de uma ferramenta de registro de produtos criada pela empresa GS1, composta por uma sequência numérica e uma representação de barras espessas e diferentes posicionamentos. O código é lido por um leitor, transformando em números binários compreensíveis ao computador, tal leitura poderá ser feita até mesmo com o código de cabeça para baixo, de forma horizontal, identificando assim o produto. As barras brancas refletirão luz e serão identificadas como 0(zero), já as barras pretas, não refletirão luz e serão referenciadas como 1(um). O número representado está diretamente ligado ao número de barras pretas e brancas.

O padrão mais utilizado e, por decorrência, o utilizado nesse projeto é o EAN-13, o mesmo possui 95 barras pretas ou brancas e pode representar 13(treze) números. O modelo EAN-13 tem como característica básica sempre iniciar com a seguinte sequência de barras, preta, branca e preta, no meio do código haverá 5 barras(branca, preta, branca, preta e branca) e no final repetirá as 3 barras do início, esse 3(três) conjuntos de barras posuuirão as barras pretas mais alongadas do código. Os 3 conjuntos de barras tem como finalidade a separação de dois grupos de barras(as 84 barras restantes), sendo um lado esquerdo e um direito, ambos com 42 barras cada. Em cada lado, haverá 6 divisões de barras, cada uma com 7 barras, e cada divisão será representada por um número. Em cada lado, esquerdo ou direito, haverá formas distintas de representar os números, pois ao lado esquerdo será de paridade ímpar, já ao lado direita a paridade será par. O primeiro par esquerdo smepre se iniciará com um valor ímpar, sendo os restantes 5 pares ou ímpares. Os conjuntos da esquerda possiblitaram a representação do primeiro número, que fica ao lado esquerdo do início, por exemplo, a sequência ímpar, par, ímpar, par, ímpar e par definiram o número 7.O lado esquerdo possui duas formas de representação numérica, uma para uma quantia par de barras pretas, e outra para um ímpar, já o lado direito possui uma quantia par de barras pretas para todos os conjuntos. Por conta da diferença de paridade de ambos os lados, o leitor poderá identificar o sentido de leitura, pois, se começar com um número ímpar de barras pretas ou finalizar com uma quantia par de barras pretas, esse será o sentido verdadeiro.

Os 3(três) primeiros dígitos do código referênciam o país de origem do produto, por exemplo, os valores que representam o Brasil são o 789 e o 790. Poderá haver países maiores com mais de um número. Caso o produto seja uma publicação literária, o código será 977, 978 e 979.
O último número do código de barras será o dígito verificador, sua função é verificar se o número fora lido corretamente. O computador irá multiplicar os algarismos do código por 1 e o dígito de verificação, de maneira revezada, até o número anterior ao próprio dígito de verificação, e então realizará a soma, em seguida, o sistema gera um número multiplo de 10(dez) mais próximo do valor resultado, logo em seguida realizará a subtração desses números, se o resultado da subtração for o número identificador, o significa que o código foi lido corretamente.

Os restantes 9(nove) números entre os dígitos de verificação e identificação de país/obra literária, serão divididos em 2(dois) grupos de tamanhos variáveis. O primeiro define o código do fabricante, e o segundo o código do produto(caso a empresa possua até 999 produtos, necessitará apenas de 3 dígitos). A quantia de dígitos usados está ligado a uma taxa paga pelas empresas à GS1.
O produto cadastrado será armazenado em um servidor da GS1, e poderá ser lido em qualquer lugar do mundo.

### 3.3 Criar sequência de números base 10


####3.3.1 Concatenar dígito verificador

Inicialmente, pensamos nas funções responsáveis pela construção da sequência de dígitos (na base decimal) que irão constituir o código de barras, sendo estes: dígito verificador, código do país, código da empresa e código do produto.



Abaixo, são descritas as funções *converter_numero*, utilizada para receber do usuário os dois primeiros dígitos do código, que serão responsáveis pelo formato dos próximos dígitos; e a função validar entrada, que confirma que os dados inseridos pelo usuários são válidos para o propósito da aplicação.

```python
def converter_numero(num):
    if num == 0:
        return "00"
    elif num == 1:
        return "01"
    else:
        return None  # Retorna None se o número não for 0 ou 1

# Função para validar a entrada do usuário
def validar_entrada(texto):
    while True:
        entrada = input(texto)
        if entrada in ["00", "01"]:
            return int(entrada)
        else:
            print("Por favor, insira apenas 00 ou 01.")

# Pedindo ao usuário para inserir um número
entrada = validar_entrada("Por favor, insira 00 ou 01: ")

# Convertendo o número usando a função converter_numero
inicio_numeracao = str(converter_numero(entrada))
```

```python
def obter_codigo(nome_pais, paises):
    for codigo, nome in paises.items():
        if nome.lower() == nome_pais.lower():
            return codigo

    return None  # Retorna None se o país não estiver na biblioteca

#A função espera um nome de um país
while True:
    pais = input("Por favor, insira o nome do país: ")
    codigo = obter_codigo(pais, paises)
    if codigo is not None:
        print("Código correspondente:", codigo)
        codigo = str(codigo)
        segunda_parte_numeracao = inicio_numeracao + codigo
        print("Número completo:", segunda_parte_numeracao)
        break
    else:
        print("País não encontrado na biblioteca. Tente novamente.")
```

###3.4 Gerar código de barras

####3.4.1 Traduzir para 0s e 1s


Tendo o código numérico dividido em três partes, criou-se a função *codificar_codigo_barras* para codificar o código de barras transformando-o de números em base 10 para uma outra representação em base 2, que será utilizada para criar as barras pretas e brancas.

Esta função usa como parâmetros a parte do código original que será codificada e qual dicionário de códigos será utilizado na tradução.

```python
def codificar_codigo_barras(numero_original, codigo_dict):
    # Lista para armazenar a sequência de barras codificada
    seq_barras = []

    # Dividir o número original em dígitos individuais
    digitos = [int(digito) for digito in numero_original]

    # Iterar sobre cada dígito e encontrar a sequência de barras correspondente
    for digito in digitos:
        # Verificar se o dígito está presente no dicionário
        if str(digito) in codigo_dict.values():
            for padrao, valor in codigo_dict.items():
                if valor == str(digito):
                    seq_barras.append(padrao)
                    break
        else:
            # Caso contrário, o dígito não está presente no dicionário
            print(f"Erro: Dígito {digito} não encontrado no dicionário.")

    # Juntar as sequências de barras para formar a sequência completa
    codigo_barras = "".join(seq_barras)

    return codigo_barras
```

####3.4.2 Gerar código de barras a partir da última numeração criada


Para isso, foi criada a função *cria_cb*, que usa como parâmetro a sequência de números zero e um que foi construída, separando-a dígito por dígito dentro de um novo array e inserindo os grupos de barras "010" nas extremidades.

Depois disso, o array é plotado utilizando a biblioteca matplotlib e a imagem correspondente ao plot é salva num arquivo.

```python
def cria_cb(numeracao_bin):

    # Criar uma lista de dígitos do número binário
    lista_b = [int(digito) for digito in numeracao_bin]

    # Criar um array numpy a partir da lista de dígitos
    array = np.array([lista_b])

    # Repetir a primeira linha 'repeticoes' vezes
    array_f = np.tile(array, (30, 1))

    # Definir as últimas 5 linhas como zeros (brancas)
    array_f[-10:] = 1

    # Adicionar "010" apenas nos extremos das últimas 5 linhas, repetindo 5 vezes
    for _ in range(5):
        for i in range(20, 25):
            array_f[i][0] = 0
            array_f[i][2] = 0
            array_f[i][84] = 0
            array_f[i][82] = 0
            if array_f[18][43] == 0:
                array_f[i][43] = 0
            if array_f[18][42] == 0:
                array_f[i][42] = 0
            if array_f[18][44] == 0:
                array_f[i][44] = 0


    # Exibir o array resultante
    plt.imshow(array_f, cmap='gray')
    plt.axis('off')
    plt.savefig('codigo_de_barras', bbox_inches='tight', pad_inches=0.1, dpi=100)
    plt.show()
    print("\nValor original utilizado para criar o codigo de barras:")
    print(numeracao_bin)

base_para_criar_codigo_de_barras = final

cria_cb(base_para_criar_codigo_de_barras)
```

###3.5 Decodificar código de barras

####3.5.1 Decodificar código de barras gerado (0s e 1s)

A função *decodificar_codigo_barras* é utilizada para este propósito, adotando como parâmetro a sequência de barras (representada por números 0 e 1) e em seguida calculando o dígito de comando.

O próximo passo é remover os grupos de barras auxiliares nas extremidades e dividir as barras restantes em grupos de sete, para que cada grupo possa ser percorrido por um laço e comparado com as sequências de dígitos armazenadas nos dicionários, a fim de descobrir qual algarismo em base 10 ele representa.

Cada algarismo correspondente é, então, armazenado num outro array que é depois atribuído a outra variável e concatenado ao dígito de comando. Dessa forma, a função retorna o código de barras decodificado.

```python
def decodificar_codigo_barras(seq_barras):

    # Lista para armazenar os dígitos decodificados
    digitos = []

    # Ignorar os três primeiros e os três últimos caracteres da sequência de barras e guardar o digito para comparação
    digito_de_comando = seq_barras[3] + seq_barras[4]
    seq_barras = seq_barras[5:-3]

    # Divida a sequência de barras em grupos de 7
    grupos = [seq_barras[i:i+7] for i in range(0, len(seq_barras), 7)]

    # Iterar sobre cada grupo de barras e encontrar o dígito correspondente
    for grupo in grupos:
        # Verifique se a sequência de barras está presente no dicionário G
        if grupo in esquerda_00:
            digitos.append(esquerda_00[grupo])
        # Verifique se a sequência de barras está presente no dicionário L
        elif grupo in direita:
            digitos.append(direita[grupo])
        # Verifique se a sequência de barras está presente no dicionário R
        elif grupo in esquerda_01:
            digitos.append(esquerda_01[grupo])
        else:
            # Caso contrário, o padrão de barras não corresponde a nenhum dígito conhecido
            digitos.append("?")

    # Junte os dígitos decodificados para formar o número original
    numero_original = "".join(digitos)

    numero_original = digito_de_comando + numero_original

    return numero_original

# Teste
codigo_barras = base_para_criar_codigo_de_barras
decodificado = decodificar_codigo_barras(codigo_barras)
print("Número Original: ",seq_inicial,"\nNúmero traduzido:", decodificado)
```

####3.5.3 Traduzir o valor encontrado da linha do array

Agora faz-se necessário traduzir essa sequência de zeros e uns para uma outra sequência que represente cada barra com somente um dígito zero ou um ao invés de 5, como é o caso da sequência inicial gerada através da leitura da imagem.

Isto é feito por meio de um laço que percorre cada dígito presente na sequência inicial e realiza uma contagem que funciona da seguinte forma: toda vez que o laço conta um número 1, ele zera o contador de números 0 e adiciona +1 ao contador de  números 1; e toda vez que o laço conta 5 números 1 seguidos, por exemplo, ele zera o contador de números 1 e adiciona o dígito "1" à sequência que está sendo gerada. O processo equivalente ocorre com a contagem dos números 0.

```python
def recupera_numeracao(sequencia):
    # Remove os 3 primeiros números e os dois últimos
    resultado = ""
    contador_zeros = 0
    contador_uns = 0

    for digito in sequencia:
        if digito == '0':
            contador_uns = 0  # Zera o contador de uns
            contador_zeros += 1
            if contador_zeros == 5:
                resultado += '0'
                contador_zeros = 0  # Zera o contador de zeros
        elif digito == '1':
            contador_zeros = 0  # Zera o contador de zeros
            contador_uns += 1
            if contador_uns == 5:
                resultado += '1'
                contador_uns = 0  # Zera o contador de uns
        else:
            contador_zeros = 0  # Zera o contador de zeros
            contador_uns = 0  # Zera o contador de uns
    resultado = resultado[2:-1]
    return resultado

# Exemplo de uso:
sequencia = linha_string
resultado = recupera_numeracao(sequencia)
print("Sequência Original:  ", base_para_criar_codigo_de_barras)
print("Sequência resultante:", resultado)
print("Número Original da base do projeto:",seq_inicial,"\nNúmero traduzido:", decodificar_codigo_barras(resultado))
```

#### 3.6.1 Gera uma imagem modificada (para testes com processamento de imagens)



O código abaixo (função *cria_cb_gradiente*) gera uma imagem modificada, neste caso, com um gradiente, com a finalidade de realizar alguns testes envolvendo processamento de imagens.

```python
#Gera a mesma imagem só que com gradiente na escala de cinza
def cria_cb_gradiente(numeracao_bin):
    # Criar uma lista de dígitos do número binário
    lista_b = [int(digito) for digito in numeracao_bin]

    # Criar um array numpy a partir da lista de dígitos
    array = np.array([lista_b])

    # Repetir a primeira linha 'repeticoes' vezes
    array_f = np.tile(array, (30, 1))

    # Criar um gradiente
    gradiente = np.linspace(0, 1, array_f.shape[1])

    # Criar uma matriz com o gradiente nas linhas
    gradiente_matrix = np.vstack([gradiente for _ in range(array_f.shape[0])])

    # Combinar o gradiente com o array_f
    array_f = array_f + gradiente_matrix

    # Definir as últimas 5 linhas como zeros (brancas)
    array_f[-10:] = 1

    # Adicionar "010" apenas nos extremos das últimas 5 linhas, repetindo 5 vezes
    for _ in range(5):
        for i in range(20, 25):
            array_f[i][0] = 0
            array_f[i][2] = 0
            array_f[i][84] = 0
            array_f[i][82] = 0

    # Exibir o mesmom array resultante acima
    plt.imshow(array_f, cmap='gray')
    plt.axis('off')
    plt.savefig('codigo_de_barras_gradiente', bbox_inches='tight', pad_inches=0.1, dpi=100)
    plt.title('Imagem modificada')
    plt.show()
    print("\nValor original utilizado para criar o codigo de barras:")
    print(numeracao_bin)

base_para_criar_codigo_de_barras = final

cria_cb_gradiente(base_para_criar_codigo_de_barras)
```

#### 3.6.2 Tratando a imagem modificada

O tratamento da imagem gerada anteriormente aconteceu por meio da função abaixo (*binarizar_imagem*) que recebe uma imagem e um valor de limiar. A funçao (*img_tratada*) é responsável por recortar a imagem, trazendo apenas as informações necessárias. O intuito de escolher uma imagem com gradiente foi mostra que, independe da intensidade da coloração das barras, é possível tratar a imagem de maneira satisfatória. Essas modificações ficam bem perceptíveis nas imagens ilustrativas.

```python
def img_tratada(imagem):
    # Detectar bordas na imagem
    aresta = cv.Canny(imagem, 50, 150, apertureSize=3)

    # Procurar contornos na imagem
    contorno, _ = cv.findContours(aresta, cv.RETR_EXTERNAL, cv.CHAIN_APPROX_SIMPLE)

    # Encontrar o menor retângulo delimitador que envolve todas as barras
    x, y, w, h = cv.boundingRect(np.concatenate(contorno))

    # Pegar só a região destcada
    regiao_destacada = imagem[y:y+h, x:x+w]

    return regiao_destacada

def binarizar_imagem(imagem, limiar):
    # Aplicar a binarização usando o limiar especificado
    _, img_binarizada = cv.threshold(imagem, limiar, 255, cv.THRESH_BINARY)

    return img_binarizada

# Caminho da imagem
caminho_imagem = "/content/codigo_de_barras_gradiente.png"

# Carregar a imagem
img_original = cv.imread(caminho_imagem, cv.IMREAD_GRAYSCALE)

# Binarizar a imagem
img_binarizada = binarizar_imagem(img_original, 127)

# Processar a imagem binarizada
img_binarizada = img_tratada(img_binarizada)

# Exibir a imagem Original
plt.imshow(img_original, cmap='gray')
plt.axis('off')
plt.title('Imagem Original')
plt.show()

# Exibir a imagem tratada
plt.imshow(img_binarizada, cmap='gray')
plt.axis('off')
plt.title('Imagem Tratada')
plt.show()


```

####3.6.3 Projetar o Array em 3d

Para fins de visualização e melhor compreensão da imagem com a qual estamos trabalhando, assim como explorar e demonstrar a ampla gama de funcionalidades que podemos obter com o processsamento de imagens, foi criada uma função que plota o array do código de barras em 3D.

Utilizando, na função *array_3d*, a biblioteca matplotlib, criou-se uma visualização em 3D do array que compõe a imagem do código de barras onde podemos visualizar de outra perspectiva a variação de valor nos pixels da imagem.

```python
#Projetar uma unica linha do array em 3d
def array_3d(array):
  # Exibir o array
    print("Array:")
    print(array)

    # Exibir o array em 3D usando barras
    fig = plt.figure(figsize=(8, 8))
    ax = fig.add_subplot(111, projection='3d')

    # Obter as dimensões do array
    x, y = array.shape

    # Criar coordenadas para os pontos
    xpos, ypos = np.meshgrid(np.arange(x), np.arange(y), indexing="ij")
    xpos = xpos.ravel()
    ypos = ypos.ravel()
    zpos = np.zeros_like(xpos)

    # Tamanho das barras
    dx = 1
    dy = 0.5
    # Definir a altura das barras
    altura_das_barras = 260  # Ajuste conforme necessário
    dz = altura_das_barras * array.ravel()  # Ajuste a altura das barras conforme necessário

    ax.bar3d(xpos, ypos, zpos, dx, dy, dz, shade=True, color="lightgray", edgecolor="black")

    # Definir os rótulos dos eixos
    ax.set_xlabel('X')
    ax.set_ylabel('Comprimento')
    ax.set_zlabel('Z') #Coloracao Preto ou Branco

    ax.set_zlim(0, 300)  # Definindo os limites do eixo Z de 0 a 10
    ax.set_title('Projeção das Barras do Código em 3d')

    plt.show()
```
